```python
def asignar_posicion(index):
    if 0 <= index < 4:
        return 'Portero'
    elif 4 <= index < 21:
        return 'Defensa'
    elif 21 <= index < 33:
        return 'Mediocentro'
    elif 33 <= index < 48:
        return 'Delantero'
    else:
        return 'Desconocido'

# Aplicar la función a la nueva columna 'Posicion'
Almeria['Posicion'] = Almeria.index.to_series().apply(asignar_posicion)
```

```python
def obtener_datos_equipo(url):
    try:
        # Inicializar el driver de Chrome
        driver = webdriver.Chrome()

        # Acceder a la página web del equipo
        driver.get(url)

        # Hacer clic en el botón con el xpath proporcionado
        boton_xpath = '//*[@id="main-container"]/div[3]/div/div[1]/div[2]/div[1]/div/a'
        boton_elemento = driver.find_element(By.XPATH, boton_xpath)
        boton_elemento.click()

        # Encontrar la tabla y extraer los datos
        tabla_xpath = '//div[3]/div/div[1]/div[2]/div[1]/div/table/tbody'
        tabla_elemento = driver.find_element(By.XPATH, tabla_xpath)

        filas = tabla_elemento.find_elements(By.TAG_NAME, "tr")
        data = []

        for fila in filas:
            columnas = fila.find_elements(By.TAG_NAME, "td")
            fila_data = [columna.text for columna in columnas]

            # Extraer el atributo alt de la imagen en la primera columna
            image_element_xpath = './/td[1]/div/a/img'
            try:
                image_element = fila.find_element(By.XPATH, image_element_xpath)
                alt_text = image_element.get_attribute("alt")
                fila_data.append(alt_text)
            except:
                pass

            data.append(fila_data)

        # Cerrar el driver después de obtener los datos
        driver.quit()

        # Verificar si hay datos antes de crear el DataFrame
        if data:
            # Convertir los datos en un DataFrame
            columnas = ["Columna1", "Columna2", "Columna3", "Columna4"]  # Ajustar las columnas según sea necesario
            df_resultados = pd.DataFrame(data, columns=columnas)
            return df_resultados
        else:
            return None

    except Exception as e:
        print(f"Error al procesar la URL {url}: {e}")
        return None

# Lista para almacenar los resultados de cada URL
resultados = Parallel(n_jobs=-1)(
    delayed(obtener_datos_equipo)(url) for url in urls_equipos
)

# Filtrar los DataFrames no vacíos antes de concatenar
resultados_filtrados = [df for df in resultados if df is not None]

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(resultados_filtrados, ignore_index=True)
```
