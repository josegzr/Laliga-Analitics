```python
def asignar_posicion(index):
    if 0 <= index < 4:
        return 'Portero'
    elif 4 <= index < 21:
        return 'Defensa'
    elif 21 <= index < 33:
        return 'Mediocentro'
    elif 33 <= index < 48:
        return 'Delantero'
    else:
        return 'Desconocido'

# Aplicar la función a la nueva columna 'Posicion'
Almeria['Posicion'] = Almeria.index.to_series().apply(asignar_posicion)
```

```python
def obtener_datos_equipo(url):
    try:
        # Inicializar el driver de Chrome
        driver = webdriver.Chrome()

        # Acceder a la página web del equipo
        driver.get(url)

        # Hacer clic en el botón con el xpath proporcionado
        boton_xpath = '//*[@id="main-container"]/div[3]/div/div[1]/div[2]/div[1]/div/a'
        boton_elemento = driver.find_element(By.XPATH, boton_xpath)
        boton_elemento.click()

        # Encontrar la tabla y extraer los datos
        tabla_xpath = '//div[3]/div/div[1]/div[2]/div[1]/div/table/tbody'
        tabla_elemento = driver.find_element(By.XPATH, tabla_xpath)
        columna1 = []
        columna2 = []
        columna_imagen = []
        
        filas = tabla_elemento.find_elements(By.TAG_NAME, "tr")
        
        for fila in filas[1:]:  # Exclude the first row (index 0)
            # Obtener datos de la primera columna
            dato_columna1_xpath = './/td[1]/div/div[2]/p/a'
            try:
                dato_columna1 = fila.find_element(By.XPATH, dato_columna1_xpath).text
                columna1.append(dato_columna1)
            except Exception as e:
                print(f"Error obteniendo Columna1: {e}")
                columna1.append(None)

            # Obtener datos de la segunda columna
            dato_columna2_xpath = './/td[2]'
            try:
                dato_columna2 = fila.find_element(By.XPATH, dato_columna2_xpath).text
                columna2.append(dato_columna2)
            except Exception as e:
                print(f"Error obteniendo Columna2: {e}")
                columna2.append(None)

            # Obtener datos de la columna de la imagen
            image_element_xpath = './/td[1]/div/a/img'
            try:
                image_element = fila.find_element(By.XPATH, image_element_xpath)
                alt_text = image_element.get_attribute("alt")
                columna_imagen.append(alt_text)
            except Exception as e:
                print(f"Error obteniendo ColumnaImagen: {e}")
                columna_imagen.append(None)

        # Crear un DataFrame con los datos recopilados
        data = {'Columna1': columna1, 'Columna2': columna2, 'ColumnaImagen': columna_imagen}
        df = pd.DataFrame(data)

        return df  # Devolver el DataFrame

    except Exception as e:
        print(f"Error al procesar la URL {url}: {e}")
        return None
    finally:
        driver.quit()  # Asegúrate de cerrar el driver en todos los casos

# Lista para almacenar los resultados de cada URL
resultados = Parallel(n_jobs=-1)(
    delayed(obtener_datos_equipo)(url) for url in urls_equipos
)

# Filtrar los DataFrames no vacíos antes de concatenar
resultados_filtrados = [df.iloc[0:] for df in resultados if df is not None]

# Concatenar todos los DataFrames en uno solo
goleadores = pd.concat(resultados_filtrados, ignore_index=True)
goleadores.head()
```
